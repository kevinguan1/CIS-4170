```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Bronx Data Set.csv", )

Bronx = df.plot(kind='bar', x="Crime Complaints", y=["2024"], title='Bronx Crime Rates')

# Rotate x-axis labels for readability
plt.xticks(rotation=25)

# Annotate bars with values
for bars in Bronx.containers:
    Bronx.bar_label(bars, fmt='%.0f', label_type='edge', fontsize=10, padding=3)

# Customize the plot title
plt.title('Bronx Crime Rates in 2024', fontsize=16, fontweight='bold')

# Add y-axis title
plt.ylabel('Number of Crime Rates')

# Adjust y-axis limit to give more space
plt.ylim(0, df["2024"].max() * 1.1)

plt.show()
```
